```python
# IMPORTANDO BIBLIOTECAS
import cv2 as cv
import numpy as np
import os
from google.colab.patches import cv2_imshow
```

```python
# Definindo diretoria da pasta de imagens
pasta_imagens = "/content/drive/MyDrive/Colab Notebooks/VS COMP/Terceira Avaliacao/Criador de Mosaico/faces"
```

```python
# Listando os arquivos em uma variavel
filenames = os.listdir(pasta_imagens)
```

```python
# Carregando imagens
imagens = [] # vetor vazio
# Montar o loop para preencher o vetor imagem
for filename in filenames:
  imagem = cv.imread(os.path.join(pasta_imagens, filename))
  imagens.append(imagem)
```

```python
# TAMANHO DAS IMAGENS
tamanho_imagem = 250
```

```python
# TAMANHO DO MOSAICO
tamanho_mosaico = 8
```

```python
# CARREGANDO AS IMAGENS DO VETOR
imagens = np.array(imagens)
```

```python
# Validar o tamanho das imagens no vetor
if len(imagens) != tamanho_mosaico ** 2:
  raise ValueError(f"Número de imagens ({len(imagens)}) não é compatível com mosaico {tamanho_mosaico}x{tamanho_mosaico}")
```

```python
# Criar o mosaico 8 x 8 com as imagens do vetor
mosaico = np.zeros((tamanho_mosaico * tamanho_imagem, tamanho_mosaico * tamanho_imagem, 3), dtype=np.uint8)
```

```python
# Montando mosaico com as imagens em loop
# Posicionar as imagens no mosaico
linha = 0
coluna = 0
for i in range(len(imagens)):
  mosaico[linha:linha + tamanho_imagem, coluna:coluna + tamanho_imagem] = imagens[i]
  coluna += tamanho_imagem
  if coluna >= tamanho_mosaico * tamanho_imagem:
    coluna = 0
    linha += tamanho_imagem
#
cv.waitKey(0)
cv.destroyAllWindows()
#Salvando imagem
cv.imwrite("mosaico2.jpg", mosaico) # moisaco 2 é o nome do arquivo a ser salvo na secao Arquivos do Colab
```
